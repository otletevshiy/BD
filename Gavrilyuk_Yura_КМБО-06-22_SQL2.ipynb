# Задание 1

```python
%%sql
SELECT
    g.Name AS GroupName,
    t.Name AS SongName,
    a.Title AS AlbumTitle,
    company
FROM
    tracks t, customers
JOIN
    albums a ON t.AlbumId = a.AlbumId
JOIN
    artists g ON a.ArtistId = g.ArtistId
JOIN
    genres ge ON t.GenreId = ge.GenreId
WHERE
    ge.Name = 'Rock' and company = 'Microsoft Corporation'
ORDER BY
    g.Name, a.Title, t.Name;
```

# Задание 2

```python
%%sql
SELECT
    t.Name AS TrackName,
    a.Name AS ArtistName
FROM
    tracks t
JOIN
    albums al ON t.AlbumId = al.AlbumId
JOIN
    artists a ON al.ArtistId = a.ArtistId
JOIN
    genres g ON t.GenreId = g.GenreId
WHERE
    g.Name = 'Blues'
    AND t.MediaTypeID IN (SELECT MediaTypeId FROM media_types WHERE Name = 'MPEG audio file')
ORDER BY
    t.Bytes DESC
LIMIT 10;
```

# Задание 3

```python
%%sql
SELECT
    c.Company AS CompanyName,
    COUNT(t.TrackId) AS TotalJazzTracks
FROM
    customers c 
JOIN
    invoices i ON c.CustomerId = i.CustomerId 
JOIN
    invoice_items ii ON i.InvoiceId = ii.InvoiceId 
JOIN
    tracks t ON ii.TrackId = t.TrackId
JOIN
    genres g ON t.GenreId = g.GenreId
WHERE
    g.Name = 'Jazz' and c.Company is not null
GROUP BY
    c.Company
ORDER BY
    TotalJazzTracks DESC;
```

# Задание 4

```python
%%sql
SELECT
    a.Title AS AlbumTitle,
    SUM(t.Milliseconds) / 60000 AS AlbumDurationMinutes
FROM
    albums a
JOIN
    tracks t ON a.AlbumId = t.AlbumId
GROUP BY
    a.AlbumId
ORDER BY
    AlbumDurationMinutes DESC
LIMIT 3;
```

# Задание 5

```python
%%sql
SELECT
    g.Name AS GenreName,
    mt.Name AS MediaType,
    ROUND(AVG(t.UnitPrice), 2) AS AveragePrice,
    COUNT(t.TrackId) AS TotalTracks
FROM
    tracks t
JOIN
    genres g ON t.GenreId = g.GenreId
JOIN
    media_types mt ON t.MediaTypeId = mt.MediaTypeId
WHERE
    t.UnitPrice > 1.5 
GROUP BY
    g.Name, mt.Name
HAVING
    COUNT(t.TrackId) > 0
    AND SUM(CASE 
            WHEN t.UnitPrice <= 1.5 THEN 1 
            ELSE 0 end) = 0
```
